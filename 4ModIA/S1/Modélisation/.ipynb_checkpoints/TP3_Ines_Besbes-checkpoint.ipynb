# 1. Introduction

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
L’objectif de ce TP est d’étudier l’action de limiteurs classiques sur les schémas de Beam-Warming et Lax-Wendroff. 
Le but est de trouver le limiteur qui permet de rendre le schéma le plus monotone possible. Cela permettra d'avoir des schémas plus stables et plus exacts.

</div>

# 2. Implémentation des schémas, de la condition initiale et de la fonction de variation totale de $u^n$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2.1 Paramètres physiques, condition initiale créneau

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
On construit dans un premier temps le maillage selon la longueur du domaine $L$ et le nombre de points de discrétisation $m$. <br><br>

Dans ce TP, on va travailler avec la condition initiale créneau, notée CI2, qui représente une discontinuité dans le champ de $u$ à la position $x = x_0$. <br>

Afin d'implémenter nos schémas, on utilise le slicing au lieu des boucles for. C'est plus optimal et plus rapide en terme d'exécution. <br>

On tient également compte des conditions limites périodiques en espace : $ u(0, t) = u(L, t) \quad\forall t $.

</div>

```python
a = 1  # Vitesse de convection en m/s
L = 1  # Longueur du domaine en m
m = 50  # Nombre de points de discrétisation
x0 = 0.5  # Position pour la condition initiale CI2
T = 1  # Temps final en s
C = 0.5  # Nombre de Courant
x = np.linspace(0,L,m) # Construction du maillage
```

```python
#Condition initiale créneau
def CI2(x, m):
    X = np.zeros(m)
    for i in range(len(x)) :
        if x[i]<x0:
            X[i] = 0
        else:
            X[i] = 1
    return X
```

## 2.2 Schéma de Beam-Warming

```python
def BW(Un, C, m, psi):
    U = np.zeros(m)
    r = np.zeros(m)
    Psi = np.zeros(m)
    r[0] = (Un[1]-Un[0])/max(abs(Un[0]-Un[-1]),1e-9)*np.sign(Un[0]-Un[-1])
    r[1] = (Un[2]-Un[1])/max(abs(Un[1]-Un[0]),1e-9)*np.sign(Un[1]-Un[0])
    r[m-1] = (Un[0]-Un[m-1])/max(abs(Un[m-1]-Un[m-2]),1e-9)*np.sign(Un[m-1]-Un[m-2])

    for i in range(m-3):
        r[2+i] = (Un[3+i]-Un[2+i])/max(abs(Un[2+i]-Un[1+i]), 1e-9)*np.sign(Un[2+i]-Un[1+i])

    for i in range(m):
        Psi[i] = psi(r[i])

    U[0] = Un[0] - C*(Un[0]-Un[-1]) - C/2*(1-C)*Psi[0]*(Un[0]-Un[-1]) + C/2*(1-C)*Psi[-1]*(Un[-1] - Un[-2])
    U[1] = Un[1] - C*(Un[1]-Un[0]) - C/2*(1-C)*Psi[1]*(Un[1]-Un[0]) + C/2*(1-C)*Psi[0]*(Un[0] - Un[-1])
    U[m-1] = Un[m-1] - C*(Un[m-1]-Un[m-2]) - (C/2)*(1-C)*Psi[m-1]*(Un[m-1]-Un[m-2]) + C/2*(1-C)*Psi[m-2]*(Un[m-2] - Un[m-3])
    U[2:m-1] = Un[2:m-1] - C*(Un[2:m-1]-Un[1:m-2]) - C/2*(1-C)*Psi[2:m-1]*(Un[2:m-1]-Un[1:m-2]) + C/2*(1-C)*Psi[1:m-2]*(Un[1:m-2] - Un[0:m-3])

    return U
```

## 2.3 Schéma de Lax-Wendroff

```python
def LW(Un, C, m, psi):
    U = np.zeros(m)
    R = np.zeros(m)
    Psi = np.zeros(m)
    
    R[0] = (Un[0]-Un[-1])/max(abs(Un[1]-Un[0]),1e-9)*np.sign(Un[1]-Un[0])
    R[1] = (Un[1]-Un[0])/max(abs(Un[2]-Un[1]),1e-9)*np.sign(Un[2]-Un[1])
    R[m-1] = (Un[m-1]-Un[m-2])/max(abs(Un[0]-Un[m-1]),1e-9)*np.sign(Un[0]-Un[m-1])

    for i in range(m-3):
        R[2+i] = (Un[2+i]-Un[1+i])/max(abs(Un[3+i]-Un[2+i]), 1e-9)*np.sign(Un[3+i]-Un[2+i])

    for i in range(m):
        Psi[i] = psi(R[i])

    U[0] = Un[0] - C*(Un[0]-Un[-1]) - C/2*(1-C)*Psi[0]*(Un[1]-Un[0]) + C/2*(1-C) *Psi[-1]*(Un[0] - Un[-1])
    U[1] = Un[1] - C*(Un[1]-Un[0]) - C/2*(1-C)*Psi[1]*(Un[2]-Un[1]) + C/2 *(1-C)*Psi[0]*(Un[1] - Un[0])
    U[m-1] = Un[m-1] - C*(Un[m-1]-Un[m-2]) - C/2*(1-C)*Psi[m-1]*(Un[0]-Un[m-1]) + C/2 *(1-C)*Psi[m-2]*(Un[m-1] - Un[m-2])
    U[2:m-1] = Un[2:m-1] - C*(Un[2:m-1]-Un[1:m-2]) - C/2*(1-C)*Psi[2:m-1]*(Un[3:]-Un[2:m-1]) + C/2*(1-C) *Psi[1:m-2]*(Un[2:m-1] - Un[1:m-2])

    return U
```

## 2.4 Variation totale de $u^n$

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
On va à présent implémenter une fonction <b>TV</b> permettant de calculer la variation totale de $u^n$ en fonction du temps. Cette fonction est utilisée afin d'évaluer la qualité des solutions numériques au fil du temps.
</div>

```python
def TV(u,m):
    res = 0
    for i in range(m):
        if i ==m-1:
            res += abs(u[0]-u[-1])
        else:
            res += abs(u[i+1]-u[i])
    return res
```

# 3. Implémentation des limiteurs 

## 3.1 Min-mod

```python
def MIN_MOD(r):
    if r>=0:
        return min(1,r)
    else:
        return 0
```

## 3.2 Superbee

```python
def SUPERBEE(r):
    return max(0, min(2*r, 1), min(r, 2))
```

## 3.3 Van-Leer

```python
def VAN_LEER(r): 
    return (r+abs(r))/(1+r)
```

## 3.4 Van-Albada

```python
def VAN_ALBADA(r):
    return (r**2+r)/(r**2+1)
```

## 3.5 Sans Limiteur 

```python
def SANS_LIMITEUR(r):
    return 1
```

# 4. Visualisation des solutions

```python
def visu_schemas(limiteur):
    dt = C / (m * a)
    u1 = CI2(x, m)
    u2 = CI2(x, m)
    time = 0.0
    var1 = [TV(u1, m)]
    var2 = [TV(u1, m)]
    t = [time]

    while time < T:
        u1 = LW(u1, C, m, limiteur)
        u2 = BW(u2, C, m, limiteur)
        var1.append(TV(u1, m))
        var2.append(TV(u2, m))
        time += dt
        t.append(time)

    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

    ax1.plot(x, CI2(x, m), label='Solution initiale')
    ax1.plot(x, u1, label='Solution finale')
    ax2.plot(x, CI2(x, m), label='Solution initiale')
    ax2.plot(x, u2, label='Solution finale')
    
    if limiteur == SANS_LIMITEUR:
        ax1.set_title("LW sans limiteur")
        ax2.set_title("BW sans limiteur")
    else:
        ax1.set_title(f'LW avec le limiteur {limiteur.__name__}')
        ax2.set_title(f'BW avec le limiteur {limiteur.__name__}')
        
    ax1.set_xlabel('x')
    ax1.set_ylabel('u')
    ax1.legend()
    ax2.set_xlabel('x')
    ax2.set_ylabel('u')
    ax2.legend()

    plt.tight_layout()

    plt.show()

    return t, var1, var2
```

## 4.1 Solutions sans limiteur

```python
t, LW_sans_lim, BW_sans_lim = visu_schemas(SANS_LIMITEUR)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que les schémas ne sont pas monotones, suite à la présence de petites oscillations dans les extremas. On a également l'impression que les deux schémas se ressemblent fortement, et diffèrent seulement d'un point de vue géométrique.
</div>

## 4.2 Solutions avec limiteur

```python
t, LW_minmod, BW_minmod = visu_schemas(MIN_MOD)
```

```python
t, LW_sup, BW_sup = visu_schemas(SUPERBEE)
```

```python
t, LW_VL, BW_VL = visu_schemas(VAN_LEER)
```

```python
t, LW_VA, BW_VA = visu_schemas(VAN_ALBADA)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que les schémas deviennent beaucoup plus lisses avec la présence de limiteurs. La solution ne fait plus apparaître aucun nouvel extremum. Le schéma est plus stable et la solution est plus exacte. 
</div>

# 5. Visualisation de l'évolution temporelle de $u^n$

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
ax1.plot(LW_minmod, label="Min Mod")
ax1.plot(LW_sup, label="Superbee")
ax1.plot(LW_VL, label="Van Leer")
ax1.plot(LW_VA, label="Van Albada")
ax2.plot(BW_minmod, label="Min Mod")
ax2.plot(BW_sup, label="Superbee")
ax2.plot(BW_VL, label="Van Leer")
ax2.plot(BW_VA, label="Van Albada")
ax1.set_title("Lax-Wendroff : Variation totale de $u^n$ en fonction du temps")
ax1.set_xlabel('$t$')
ax1.set_ylabel('$u^n$')
ax1.legend()
ax2.set_title("Beam-Warming : Variation totale de $u^n$ en fonction du temps")
ax2.set_xlabel('$t$')
ax2.set_ylabel('$u^n$')
ax2.legend()

plt.tight_layout()

plt.show()
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Dans les deux schémas, et plus précisément avec les limiteurs Min Mod et Van Albada, on remarque que l'écart de $u^n$ augmente légèrement avec le temps, mais reste minime. La décroissance de $u^n$ témoigne de l'effet des limiteurs visant à rendre le schéma plus monotone. <br><br>

De plus, le limiteur Min Mod semble moins performant, comme le montre le fait que $u^{60}$ passe de 0.001 à 0.00062 après 60 secondes, ce qui représente une variation relativement importante par rapport aux autres limiteurs. <br>

Par ailleurs, les limiteurs les plus efficaces semblent être Van Leer et Superbee, produisant des schémas plus monotones dans l'ensemble des observations. De plus, $u^n$ est constant tout au long du temps et est à 0.001 quelque soit t.

</div>
