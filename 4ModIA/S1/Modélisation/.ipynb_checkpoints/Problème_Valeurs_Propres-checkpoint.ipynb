```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as npl
import scipy.sparse
import math
```

```python
# Definition de l'intervalle ]-R,R[ + mesh size
R = 3.0
N = 59
h = 2*R/(N+1)
X = np.linspace(-R,R,N)
```

```python
# 1D Finite Difference Matrix
def Rig(N):
    A = np.zeros([N,N])
    A = scipy.sparse.diags([-1,2,-1],[-1,0,1], shape=(N,N), format='lil').toarray()
    A = A*((N+1)**2/(4*R**2))
    #A = 2*np.eye(N,0)-1*np.eye(N-1,1)-1*np.eye(N-1,-1) 
    return A
```

```python
def U_exacte(k,x):
    U = 1/np.sqrt(R)*np.sin((k*np.pi*(x+R))/(2*R))
    norm = np.linalg.norm(U)
    return U/norm
```
