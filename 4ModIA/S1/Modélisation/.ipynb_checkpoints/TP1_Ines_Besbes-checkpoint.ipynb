# 1. Introduction

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
L'objectif de ce TP est de résoudre l'équation d'advection 1D à l'aide de plusieurs schémas de différences finies. L'équation d'advection 1D peut être représentée comme suit :

\begin{equation}
\frac{\partial u}{\partial t} + a \frac{\partial u}{\partial x} = 0 \quad \text{(1)}
\end{equation}

L'équation (1) constitue le modèle le plus simple mais néanmoins générique. Le développement de schémas numériques précis et "robustes" pour cette équation est essentiel dans la construction de schémas numériques plus complexes adaptés à la résolution des équations (ou des systèmes d'équations) de transport non linéaires intervenant en mécanique des fluides, par exemple.
</div>

# 2. Construction et visualisation du maillage et de la condition initiale

## 2.1 Importation des librairies 

```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

## 2.2 Paramètres physiques et numériques

```python
a = 1  # Vitesse de convection en m/s
L = 1  # Longueur du domaine en m
m = 50  # Nombre de points de discrétisation
x0 = 0.5  # Position pour la condition initiale CI2
T = 1  # Temps final en s
C = 0.5  # Nombre de Courant
beta = 1 #Numéro du mode considéré
```

## 2.3 Maillage et conditions initiales

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
On construit dans un premier temps le maillage selon la longueur du domaine $L$ et le nombre de points de discrétisation $m$. <br><br>

Deux conditions initiales différentes sont envisagées, notées CI1 et CI2, détaillées ci-dessous. L'objectif initial est de construire le tableau $x$, contenant les abscisses des points du maillage, et $u^0$, la solution initiale, puis de les visualiser à l'aide de Matplotlib. <br>

La première condition initiale, notée CI1, représente une évolution sinusoidale du champ de la variable $u$, où $u$ est une fonction du temps $t$ et de l'espace $x$. <br>

La deuxième condition initiale "créneau", notée CI2, représente une discontinuité dans le champ de $u$ à la position $x = x_0$.
</div>

```python
# Construction du maillage
x = np.linspace(0,1,m)
```

```python
#Conditions initiales envisagées 
def CI1(x, beta, m):
    return np.sin((2 * beta * np.pi * x) / L)

def CI2(x, beta, m):
    X = np.zeros(m)
    for i in range(len(x)) :
        if x[i]<x0:
            X[i] = 0
        else:
            X[i] = 1
    return X
```

```python
u0_CI1 = CI1(x, beta,m) # Solution initiale pour CI1
u0_CI2 = CI2(x, beta, m) # Solution initiale pour CI2
```

```python
plt.subplot(1, 2, 1)
plt.plot(x, u0_CI1, label='CI1')
plt.title('Condition Initiale 1')
plt.xlabel('Position (m)')
plt.ylabel('Amplitude')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, u0_CI2, label='CI2')
plt.title('Condition Initiale 2')
plt.xlabel('Position (m)')
plt.ylabel('Amplitude')
plt.legend()

plt.tight_layout()
plt.show()
```

# 3. Etude du comportement de schémas différences finies

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
Dans cette section, nous examinons le comportement des schémas de différences finies. <br><br>

Le **nombre de Courant $\mathcal{C}$** est défini comme suit :
$ \mathcal{C} = \frac{a \Delta t}{\Delta x} $

Ce dernier est une mesure de la stabilité des schémas de différences finies. Il est déterminé par la vitesse de convection $a$, le pas de temps $\Delta t$, et le pas d'espace $\Delta x.$ <br>

Le choix de **$\mathcal{C}$ = 0.5** par défaut s'aligne souvent avec des conditions de stabilité.

</div>

## 3.2 Implémentation des différents schémas numériques

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
Afin d'implémenter nos schémas, on utilise le slicing au lieu des boucles for. C'est plus optimal et plus rapide en terme d'exécution. <br><br>

On tient également compte des conditions limites périodiques en espace : $ u(0, t) = u(L, t) \quad\forall t $.
</div>

```python
def FOU_EE(u_n, m, C, a, L):
    dx = L / (m - 1)
    dt = C * dx / abs(a)
    u = np.zeros(m)
    u[0] = u_n[0]-a * dt / dx*(u_n[0]-u_n[-1])
    u[1:] = u_n[1:]-a * dt / dx*(u_n[1:] - u_n[:m-1])
    return u
```

```python
def FOF_EE(u_n, m, C, a, L):
    dx = L / (m - 1)
    dt = C * dx / abs(a)
    u = np.zeros(m)
    u[m-1] = u_n[m-1]-a*dt/dx*(u_n[1]-u_n[m-1])
    u[:m-1] = u_n[:m-1]-a * dt / dx*(u_n[1:] - u_n[:m-1])
    return u
```

```python
def SOC_EE(u_n, m, C, a, L):
    dx = L / (m - 1)
    dt = C * dx / abs(a)
    u = np.zeros(m)
    u[0] = u_n[0]-a * dtdef LAX_WENDROFF(u_n,m,C, a, L):
    dx = L / (m - 1)
    dt = C * dx / abs(a)
    u = np.zeros(m)
    u[0] = u_n[0]-(a * dt / dx)/2*(u_n[1]-u_n[m-1])+(a * dt / dx)**2/2*(u_n[1]-2*u_n[0]+u_n[m-1])
    u[m-1] = u_n[m-1]-(a * dt / dx)/2*(u_n[0]-u_n[m-2])+(a * dt / dx)**2/2*(u_n[0]-2*u_n[m-1]+u_n[m-2])
    u[1:m-1] = u_n[1:m-1]-(a * dt / dx)/2*(u_n[2:]-u_n[:m-2])+(a * dt / dx)**2/2*(u_n[2:]-2*u_n[1:m-1]+ u_n[:m-2])
    return u / dx/2*(u_n[1]-u_n[m-1])
    u[m-1] = u_n[m-1]-a * dt / dx/2*(u_n[0]-u_n[m-2])
    u[1:m-1] = u_n[1:m-1]-a * dt / dx/2*(u_n[2:m] - u_n[:m-2])
    return u
```

```python
def LAX_WENDROFF(u_n,m,C, a, L):
    dx = L / (m - 1)
    dt = C * dx / abs(a)
    u = np.zeros(m)
    u[0] = u_n[0]-(a * dt / dx)/2*(u_n[1]-u_n[m-1])+(a * dt / dx)**2/2*(u_n[1]-2*u_n[0]+u_n[m-1])
    u[m-1] = u_n[m-1]-(a * dt / dx)/2*(u_n[0]-u_n[m-2])+(a * dt / dx)**2/2*(u_n[0]-2*u_n[m-1]+u_n[m-2])
    u[1:m-1] = u_n[1:m-1]-(a * dt / dx)/2*(u_n[2:]-u_n[:m-2])+(a * dt / dx)**2/2*(u_n[2:]-2*u_n[1:m-1]+ u_n[:m-2])
    return u
```

## 3.3 Visualisation des solutions pour différentes valeurs de $\beta$

```python
betas = [2, 4, 8, 16]
```

```python
def visu_betas(betas, schema, a, L):
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    axes = axes.flatten()
    
    for i, beta in enumerate(betas):
        dt = C/(m*abs(a))
        u = CI1(x,beta, m)
        time = 0.0
        while time<T:
            u = schema(u, m, C, a, L)
            time += dt
    
        ax = axes[i]
    
        ax.plot(x, CI1(x,beta, m), label='Solution initiale')
        ax.plot(x, u, label='Solution finale')
        
        # ax.axis([0, 0.1, 0, 1])

        ax.set_xlabel('x')
        ax.set_ylabel('u')
        ax.set_title(f'Solution {schema.__name__} avec CI1 (Beta = {beta})')
        ax.legend()
    
    plt.tight_layout()
    plt.show()
```

### 3.3.1 Schéma FOU-EE

```python
visu_betas(betas, FOU_EE, a, L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
La variable $\beta$ affecte la fréquence de l'onde sinus. Lorsque l'on augmente la valeur de $\beta$, on observe que la solution initiale devient plus "ondulée". Ainsi, cette augmentation n'affecte pas directement le schéma FOU-EE, mais elle affecte la solution initiale qui lui est passée.

</div>

### 3.3.2 Schéma FOF-EE

```python
visu_betas(betas, FOF_EE,a, L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Lorsque l'on augmente la valeur de $\beta$, on introduit une onde sinusoidale avec une fréquence plus élevée. 
Or, on observe que plus on augmente $\beta$, et donc plus on augmente le nombre de périodes, plus il y a d'erreur. 
De plus, ce schéma utilise les valeurs des points voisins situés en amont de la direction de convection. Cela entraîne donc des phénomènes de sur-oscillation. Le schéma n'est donc pas stable.
</div>

### 3.3.3 Schéma SOC-EE 

```python
visu_betas(betas, SOC_EE,a, L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Peu importe la valeur de $\beta$, on observe des oscillations importantes, qui sont de plus en plus nombreuses quand $\beta$ est plus grand. On voit tout de même qu'il y a moins d'oscillations que sur le schéma précédent. En effet, le schéma SOC_EE, étant de second ordre, est généralement mieux équipé pour gérer ces variations rapides par rapport aux schémas de premier ordre.
</div>

### 3.3.4 Schéma de Lax-Wendroff

```python
visu_betas(betas, LAX_WENDROFF,a, L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Comme pour FOU-EE, plus on augmente $\beta$, plus la solution initiale varie. 
</div>

## 3.4 Visualisation des solutions pour différentes valeurs de C avec β fixée à 1

```python
C_array = [0.1, 0.2, 0.4, 0.8, 1, 1.5]
```

```python
def visu_C(C_array, schema, CI,a, L, beta = 1):
    fig, axes = plt.subplots(2, 3, figsize=(10, 8))
    axes = axes.flatten()

    
    for i, C in enumerate(C_array):
        dt = C/(m*abs(a))
        u = CI(x,beta, m)
        time = 0.0
        while time<T:
            u = schema(u, m, C,a, L)
            time += dt
    
        ax = axes[i]
    
        ax.plot(x, CI(x,beta, m), label='Solution initiale')
        ax.plot(x, u, label='Solution finale')
        
    
        ax.set_xlabel('x')
        ax.set_ylabel('u')
        ax.set_title(f'Solution {schema.__name__} avec {CI.__name__} (C = {C})')
        ax.legend()
    
    plt.tight_layout()
    plt.show()
```

### 3.4.1 Schéma FOU-EE

```python
visu_C(C_array, FOU_EE, CI1, a, L)
```

```python
visu_C(C_array, FOU_EE, CI2, a, L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que pour des valeurs de $\mathcal{C} \leq 1 $, le schéma reste conditionnellement stable. Cela est du au fait que le schéma numérique obtient toutes les données de l'équation physique décrivant l'état $u_i^{n+1}$. <br><br>
Toutefois, lorsque $\mathcal{C} \ge 1$, on observe que la solution finale oscille beaucoup, et le schéma devient instable. <br><br>
Dans le cas particulier où $\mathcal{C}$=1, le schéma a accès à toute l'information provenant de l'équation physique modèle. Cela conduit donc à la superposition de la solution initiale avec la solution finale. Ainsi, lorsque $\mathcal{C}$=1, le schéma numérique est capable de représenter de manière précise l'évolution de la solution au fil du temps.
</div>

### 3.4.2 Schéma FOF-EE

```python
visu_C(C_array, FOF_EE, CI1, a, L)
```

```python
visu_C(C_array, FOF_EE, CI2, a)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
En ce qui concerne schéma décentré aval d’ordre un en espace, on observe des oscillations numériques dans la solution finale quelque soit le nombre de Courant $\mathcal{C}$. <br><br>
Lorsque le schéma FOF-EE est utilisé pour résoudre des équations différentielles, il cherche toute l'information en aval, et devient instable indépendamment de la valeur du nombre de Courant. 
Cela signifie que l'augmentation du nombre de Courant $\mathcal{C}$ n'aura pas d'impact sur la stabilité du schéma, et celui-ci produira des résultats non fiables.  <br> <br>

Le schéma FOF-EE est donc inconditionellement instable.
</div>

### 3.4.3 Schéma SOC-EE

```python
visu_C(C_array, SOC_EE, CI1, a, L)
```

```python
visu_C(C_array, SOC_EE, CI2, a)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Le schéma est instable quelque soit le nombre de Courant $\mathcal{C}$.<br><br>
La taille du pas de temps doit être choisie suffisamment petite pour satisfaire une condition de stabilité, qui est liée au nombre de Courant. Lorsqu'on augmente la valeur de $\mathcal{C}$, la taille du pas de temps $\Delta t$ augmente, ce qui conduit à l'instabilité de la solution. <br><br>
Le schéma SOC-EE est donc inconditionnellement instable.
</div>

### 3.4.4 Schéma de Lax-Wendroff

```python
visu_C(C_array, LAX_WENDROFF, CI1, a, L)
```

```python
visu_C(C_array, LAX_WENDROFF, CI2, a)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Comme pour FOU-EE, le schéma est stable pour $\mathcal{C} \leq 1$, et devient instable quand $\mathcal{C} > 1$. <br><br>
Lorsque $\mathcal{C} = 1$, on voit que la solution initiale se superpose avec la solution finale. Le schéma de Lax-Wendroff est donc conditionnellement stable.

</div>

## 3.5 Visualisation des solutions avec la CI2 à différents instants T avec C=1

```python
C = 1
times = [T/8, T/4, T/2, T] 
```

```python
def visu_T(times, schema, a,L, C = 1):
    fig, axes = plt.subplots(2, 2, figsize=(10, 8))
    axes = axes.flatten()
    
    for i, tmax in enumerate(times):
        dt = C/(m*abs(a))
        u = CI2(x,beta, m)
        time = 0.0
        while time<tmax:
            u = schema(u, m, C,a,L)
            time += dt
    
        ax = axes[i]
    
        ax.plot(x, CI2(x,beta, m), label='Solution initiale')
        ax.plot(x, u, label='Solution finale')
        
    
        ax.set_xlabel('x')
        ax.set_ylabel('u')
        ax.set_title(f'Solution {schema.__name__} avec CI2 (T = {tmax}, C= {C})')
        ax.legend()
    
    plt.tight_layout()
    plt.show()
```

### 3.5.1 Schéma FOU-EE

```python
visu_T(times, FOU_EE,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Quand $T<1$, la solution finale est en décalage par rapport à la solution initiale. Elle n'atteint pas un état d'équilibre, car elle est évaluée sur une courte période de temps. 
Lorsque $T=1$, la solution finale et la solution initiale coîncident. 
</div>

### 3.5.2 Schéma FOF-EE

```python
visu_T(times, FOF_EE,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On a vu précédemment que la valeur du nombre de Courant $\mathcal{C}$ n'influait pas sur la stabilité du schéma. 
Si $\Delta x = 1$ et $a = 1$, alors la relation $T = \frac{C \cdot \Delta x}{a}$ peut être simplifiée à $T = C$.
Ainsi, peu importe la valeur de $T$, le schéma reste instable. 
</div>

### 3.5.3 Schéma SOC-EE

```python
visu_T(times, SOC_EE,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Comme pour le schéma FOF-EE, peu importe la valeur de $T$, le schéma SOC-EE reste instable.
</div>

### 3.5.4 Schéma de Lax-Wendroff

```python
visu_T(times, LAX_WENDROFF, a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Comme pour le schéma FOU-EE, quand $T<1$, la solution finale est en décalage par rapport à la solution initiale. Elle n'atteint pas un état d'équilibre, car elle est évaluée sur une courte période de temps. 
Lorsque $T=1$, la solution finale et la solution initiale coîncident. 
</div>

## 3.6 Etude de l'influence de la discrétisation spatiale avec la condition initiale CI2, avec C=0.5

```python
m_array = [4, 10, 20, 50, 100, 200]
```

```python
def visu_m(m_array, schema,a,L, C = 0.5):
    fig, axes = plt.subplots(2, 3, figsize=(10, 8))
    axes = axes.flatten()
    
    for i, m in enumerate(m_array):
        
        x = np.linspace(0,1,m)
        dt = C/(m*abs(a))
        u = CI2(x,beta, m)
        time = 0.0
        while time<T:
            u = schema(u, m, C,a,L)
            time += dt
    
        ax = axes[i]
    
        ax.plot(x, CI2(x,beta, m), label='Solution initiale')
        ax.plot(x, u, label='Solution finale')
        
    
        ax.set_xlabel('x')
        ax.set_ylabel('u')
        ax.set_title(f'{schema.__name__} avec CI2 (m = {m})')
        ax.legend()
    
    plt.tight_layout()
    plt.show()
```

### 3.7.1 Schéma FOU-EE

```python
visu_m(m_array, FOU_EE,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que plus $m$ est grand, plus la solution finale se rapproche de la solution initiale, et plus le schéma est stable. Ainsi, l'augmentation du nombre de points $m$ permet de capturer des détails plus fins dans le comportement de la solution. Cela permet également de converger vers la solution initiale.
</div>

### 3.7.2 Schéma FOF-EE

```python
visu_m(m_array, FOF_EE,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que plus $m$ est grand, plus les oscillations sont importantes, plus le schéma est instable. 
</div>

### 3.7.3 Schéma SOC-EE

```python
visu_m(m_array, SOC_EE,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Comme pour le schéma FOF-EE, plus $m$ est grand, plus il y a d'oscillations.
</div>

### 3.7.4 Schéma de Lax-Wendroff

```python
visu_m(m_array, LAX_WENDROFF,a,L)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que plus $m$ est grand, plus la solution est précise et se rapproche de la solution initiale. Les petites oscillations que l'on voit lorsque $m=200$ par exemple sont tout à fait normales, car le schéma n'est pas monotone.
Ainsi, plus la discrétisation spatiale est importante, plus le schéma est stable et convergent.
</div>

# 5. Conclusion

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
Pour conclure, les schémas FOU-EE et Lax-Wendroff sont convergents : ils sont stables et consistants. Le schéma Lax-Wendroff est d'ordre supérieur dans la précision et permet d'éviter la diffusion numérique. <br><br>

Quant aux schémas FOF-EE, il est certes simple à implémenter, mais il peut introduire de la diffusion numérique, ce qui peut altérer la précision, en particulier dans la capture de phénomènes à haute fréquence. Il est donc mieux de l'utiliser pour des problèmes simples et ne nécessitant pas une haute précision. <br>

En théorie, SOC-EE offre une précision d'ordre deux en espace, ce qui améliore la représentation des variations fines de la solution. Cependant, il peut ne pas être aussi stable que certains schémas d'ordre supérieur, nécessitant une attention particulière à la condition de stabilité numérique. 

</div>
