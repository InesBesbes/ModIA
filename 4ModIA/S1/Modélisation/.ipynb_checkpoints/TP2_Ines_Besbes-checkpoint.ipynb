# 1. Introduction

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
L'objectif de ce TP est d'examiner numériquement le comportement de schémas discrets résultant de l'application de trois méthodes d'intégration temporelle classiques (Euler explicite, RK2, Euler implicite) à des schémas semi-discrets d'ordre un (FOU) et deux (UPO2VF) décentrés en espace.
</div>

# 2. Construction et visualisation du maillage et de la condition initiale

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
On construit dans un premier temps le maillage selon la longueur du domaine $L$ et le nombre de points de discrétisation $m$. <br><br>

On va travailler avec la deuxième condition initiale "créneau", notée CI2, qui représente une discontinuité dans le champ de $u$ à la position $x = x_0$.<br>

Afin d'implémenter nos schémas numériques, on utilise le slicing au lieu des boucles for. C'est plus optimal et plus rapide en terme d'exécution. <br>

On tient également compte des conditions limites périodiques en espace : $ u(0, t) = u(L, t) \quad\forall t $.

</div>

## 2.1 Importation des librairies 

```python
import numpy as np
import matplotlib.pyplot as plt
import math 
```

## 2.2 Maillage et conditions initiales

```python
# Paramètres
a = 1  # Vitesse de convection
L = 1  # Longueur du domaine
m = 50  # Nombre de points
C = 0.5  # Nombre de Courant
T = 1  # Temps
x0 = 0.5 # Position pour la condition initiale CI2
beta = 1 # Numéro du mode considéré
```

```python
# Construction du maillage
x = np.linspace(0,L,m)
```

```python
#Condition initiale créneau
def CI2(x, m):
    X = np.zeros(m)
    for i in range(len(x)) :
        if x[i]<x0:
            X[i] = 0
        else:
            X[i] = 1
    return X
```

# 3. Schéma d'Euler explicite

## 3.1 Implémentation des schémas FOU-EE et UPO2VF

```python
def FOU_EE(u_n, m, C):
    u = np.zeros(m)
    u[0] = u_n[0]-C*(u_n[0]-u_n[-1])
    u[1:] = u_n[1:]-C*(u_n[1:] - u_n[:m-1])
    return u
```

```python
def UPO2VF(u_n, m, C):
    u = np.zeros(m)
    u[0] = u_n[0] - C/4*(u_n[1] + 3*u_n[0] -5*u_n[-1] + u_n[-2])
    u[1] = u_n[1] - C/4*(u_n[2] + 3*u_n[1] -5*u_n[0] + u_n[-1])
    u[m-1] = u_n[m-1] - C/4*(u_n[0] + 3*u_n[m-1] -5*u_n[m-2] + u_n[m-3])
    u[2:m-1] = u_n[2:m-1] - C/4*(u_n[3:] + 3*u_n[2:m-1] -5*u_n[1:m-2] + u_n[0:m-3])
    return u
```

## 3.2 Visualisation des solutions pour différents CFL

```python
C_array = [0.1, 0.5, 1]
```

```python
def visu_C(C_array, schema):
    fig, axes = plt.subplots(1, 3, figsize=(15, 5))
    axes = axes.flatten()
    
    for i, C in enumerate(C_array):
        dt = C/(m*a)
        u = CI2(x, m)
        time = 0.0
        while time<T:
            u = schema(u, m, C)
            time += dt
    
        ax = axes[i]
    
        ax.plot(x, CI2(x, m), label='Solution initiale')
        ax.plot(x, u, label='Solution finale')
        
        ax.set_xlabel('x')
        ax.set_ylabel('u')
        if schema == FOU_EE or schema == UPO2VF:
            ax.set_title(f'{schema.__name__} avec Euler Explicite (C = {C})')
        else:
            ax.set_title(f'{schema.__name__} avec RK2 Explicite (C = {C})')
        ax.legend()
    
    plt.tight_layout()
    plt.show()
```

### 3.2.1 Schéma FOU-EE

```python
visu_C(C_array, FOU_EE)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que pour des valeurs de $\mathcal{C} \leq 1 $, le schéma reste conditionnellement stable. Cela est du au fait que le schéma numérique obtient toutes les données de l'équation physique décrivant l'état $u_i^{n+1}$. <br><br>
Toutefois, lorsque $\mathcal{C} \ge 1$, on observe que la solution finale oscille beaucoup, et le schéma devient instable. <br><br>
Dans le cas particulier où $\mathcal{C}$=1, le schéma a accès à toute l'information provenant de l'équation physique modèle. Cela conduit donc à la superposition de la solution initiale avec la solution finale. Ainsi, lorsque $\mathcal{C}$=1, le schéma numérique est capable de représenter de manière précise l'évolution de la solution au fil du temps.
</div>

### 3.2.2 Schéma UPO2VF

```python
visu_C(C_array, UPO2VF)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que le schéma UPO2VF est plus sensible à la valeur de $\mathcal{C}$. En effet, on remarque que la solution finale explose en oscillation dès lors que $\mathcal{C}\geq0.5$. On passe d'une échelle à 10⁰ à une échelle 10⁹. Ainsi, le schéma n'est pas stable.
    
</div>

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Ainsi, le schéma d'Euler explicite n'est pas adapté au schéma UPO2VF.   
</div>

# 4. Schéma de Runge-Kutta explicite d’ordre deux

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
On décide à présent d'implémenter les schémas discrets combinant les discrétisations spatiales FOU et UPO2VF avec le schéma de
Runge-Kutta explicite. Ce dernier permet d'améliorer la stabilité et la précision de la méthode d’Euler en faisant
une évaluation supplémentaire des états entre $t^n$ et $t^{n+1}$.
</div>

## 4.1 Implémentation de RK2 explicite

```python
def RK2_FOU_EE(Un, m, C):
    dU = np.zeros(m)
    U = np.zeros(m)
    
    U_star = FOU_EE(Un,m,C)

    dU[1:m] = 1/2*(U_star[1:m] - U_star[0:m-1]) + 1/2*(Un[1:m] - Un[0:m-1])
    dU[0] = 1/2*((U_star[0] - U_star[-2]) + (Un[0] - Un[-2]))

    U =  Un - C*dU
    return U
```

```python
def RK2_UPO2VF(Un, m, C):
    dU = np.zeros(m)
    U = np.zeros(m)
    
    U_star = UPO2VF(Un,m,C)

    dU[2:m-1] = (U_star[3:m]+3*U_star[2:m-1]-5*U_star[1:m-2]+U_star[0:m-3])/8 + (Un[3:m]+3*Un[2:m-1]-5*Un[1:m-2]+Un[0:m-3])/8
    dU[0] = (U_star[1] + 3*U_star[0] -5*U_star[m-1] + U_star[m-2])/8 + (Un[1] + 3*Un[0] -5*Un[m-1] + Un[m-2])/8
    dU[1] = (U_star[2] + 3*U_star[1] -5*U_star[0] + U_star[m-1])/8 + (Un[2] + 3*Un[1] -5*Un[0] + Un[m-1])/8
    dU[m-1] = (U_star[0] + 3*U_star[m-1] -5*U_star[m-2] + U_star[m-3])/8 + (Un[0] + 3*Un[m-1] -5*Un[m-2] + Un[m-3])/8


    U =  Un - C*dU
    return U
```

## 4.2 Visualisation des solutions pour différents CFLs

```python
C_array = [0.1, 0.5, 1]
```

### 4.2.1 Schéma FOU-EE

```python
visu_C(C_array, RK2_FOU_EE)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe que pour des valeurs de $\mathcal{C} \leq 1$, le schéma reste conditionnellement stable. 
Toutefois, lorsque $\mathcal{C} \gt 1$, on observe que la solution finale commence à osciller et devient moins précise. <br><br>
Ainsi, le schéma RK2 explicite n'est pas bien adapté pour FOU-EE.
</div>

### 4.2.2 Schéma UPO2VF

```python
visu_C(C_array, RK2_UPO2VF)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On constate une stabilité et une précision nettement meilleures du schéma UPO2VF lorsqu'il est utilisé avec le schéma RK2 explicite par rapport au schéma d'Euler explicite. En effet, la solution finale présente une oscillation réduite et ne connaît pas de croissance exponentielle, contrairement à ce que l'on observe dans le cas d'Euler explicite, en particulier lorsque $\mathcal{C}=1$.
    
</div>

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Ainsi, le schéma RK2 explicite est plus précis à l'ordre 2 en temps, et convient donc mieux au schéma UPO2VF. 
</div>

# 5. Schéma d’Euler implicite

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
On va maintenant implémenter les schémas discrets combinant les discrétisations spatiales FOU et UPO2VF avec le schéma d'Euler implicite. Ce dernier nécéssite la résolution d'un système, et est peu précis car il est d'ordre un en temps.
</div>

## 5.1 Implémentation de RK2 implicite

```python
def FOU_EE_Imp(Un, m, C):
    v = [1+C for i in range(m)]
    v1 = [-C for i in range(m-1)]
    M = np.diag(v) + np.diag(v1,-1) 
    M[0,m-1] = -C

    dt = C / (m * a)
    step_total = int(T/dt)
    
    U = np.zeros((step_total, m))
    U[0, :] = Un

    for i in range(0, step_total-1):
        U[i+1, :] = np.linalg.solve(M, U[i, :])
        
    return U[-1, :]
```

```python
def UPO2VF_Imp(Un, m, C):
    v0 = [C/4 for i in range(m-2)]
    v1 = [-5*C/4 for i in range(m-1)]
    v2 = [1 + 3*C/4 for i in range(m)]
    v3 = [C/4 for i in range(m-1)]
    
    M = np.diag(v0,-2) + np.diag(v1,-1) + np.diag(v2) + np.diag(v3,1)
    M[0,-1] = -5*C/4
    M[0,-2] = C/4
    M[-1,0] = C/4
    M[1,-1] = C/4

    dt = C / (m * a)
    step_total = int(T/dt)
    
    U = np.zeros((step_total, m))
    U[0, :] = Un

    for i in range(0, step_total-1):
        U[i+1, :] = np.linalg.solve(M, U[i, :])
        
    return U[-1, :]
```

## 5.2 Visualisation des solutions pour différents CFLs

```python
C_array = [0.1, 0.5, 1, 10, 20]
```

```python
def visu_C_implicit(C_array, schema):
    u = CI2(x, m)
    plt.plot(x, u, label='Solution initiale')
    
    for C in C_array:
        u_new = schema(u, m, C)
        plt.plot(x, u_new, label='C='+str(C))
        
    
    plt.xlabel('x')
    plt.ylabel('u')
    plt.title(f'{schema.__name__} avec Euler Implicite')
    plt.legend()
    plt.show()
```

### 5.2.1 Schéma FOU-EE

```python
visu_C_implicit(C_array, FOU_EE_Imp)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
On observe cette fois-ci que peu importe la valeur de $\mathcal{C}$, la stabilité est assurée. On remarque que plus la valeur du nombre de Courant $\mathcal{C}$ est importante, plus la solution finale est petite. 
</div>

### 5.2.2 Schéma UPO2VF

```python
visu_C_implicit(C_array, UPO2VF_Imp)
```

<div style="background-color:rgba(255, 0,0, 0.1); padding:15px;">
Comme pour le schéma FOU, UPO2VF est stable quelque soit la valeur du nombre de Courant $\mathcal{C}$. De plus, on remarque que plus $\mathcal{C}$ est grand, moins il y a de pics dans la solution finale. 
</div>

# 6. Conclusion

<div style="background-color:rgba(0, 0, 255, 0.1); padding:15px;margin:5px">
Ainsi, le schéma décentré amont FOU-EE et le schéma UPO2VF associés à une intégration d’Euler implicite sont inconditionellement stable. Cela est du au fait que le schéma d'Euler implicite résout les équations en utilisant les valeurs au pas de temps suivant, en opposition à l'Euler explicite qui se base sur les valeurs au pas de temps courant.
