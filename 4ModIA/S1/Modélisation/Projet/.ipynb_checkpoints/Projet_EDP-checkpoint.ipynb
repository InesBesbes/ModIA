# Contexte

Une usine déverse un produit dans une rivière de largeur constante $b$ et animée d'une vitesse $V$. 

### Hypothèses physiques

On suppose que le produit est :
<li>Non miscible à l'eau</li>
<li>Non réactif avec l'eau</li>
<li>Moins dense que l'eau donc il flotte</li>

Le produit est ensuité transporté suite à un phénomène de dispersion, l'advection, traduit par l'équation suivante : 
$\begin{equation}
\frac{\partial u}{\partial t} + c \frac{\partial u}{\partial x} = 0.
\end{equation}$



On observe également un phénomène diffusif du produit dans l’eau de la rivière.




```python
import numpy as np
import numpy.linalg as npl
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import scipy.sparse
```

```python
#  Discrétisation en espace
xmin = 0.0; xmax = 1.8; nptx = 16; nx = nptx-2  
hx = (xmax-xmin)/(nptx -1)
xx = np.linspace(xmin,xmax,nptx) 
xx = xx.transpose()
xxint = xx[1:nx+1]
ymin = 0.0; ymax = 1.0; npty = 16; ny = npty-2 
hy = (ymax-ymin)/(npty -1)
yy = np.linspace(ymin,ymax,npty)
yy=yy.transpose() 
yyint = yy[1:ny+1]
mu = 1
```

```python
B = scipy.sparse.diags([-mu/hx**2, mu*(2/hy**2 + 2/hx**2), -mu/hx**2], [-1, 0, 1], shape=(nx+2,nx+2), format = 'lil').toarray()
B[0,0] = 1
B[-1,-1] = 1
B[0,1]=0
B[-1,-2]=0
# plt.imshow(B, cmap='viridis')

C = -mu/hy**2*np.eye(nx+2)
C[0,0] = 0
C[-1,-1] = 0
# plt.imshow(C, cmap='viridis')


A = np.zeros(((ny+2) * B.shape[0], (ny+2) * B.shape[1]))
A[:B.shape[0], :B.shape[1]] = np.eye(B.shape[0])

for i in range(1,(ny+2)):
    A[i * B.shape[0]:(i + 1) * B.shape[0], i * B.shape[1]:(i + 1) * B.shape[1]] = B

for i in range((ny+2) - 2):
    A[(i + 1) * B.shape[0]:(i + 2) * B.shape[0], i * B.shape[1]:(i + 1) * B.shape[1]] = C

for i in range(1,(ny+2) - 1):
    A[i * B.shape[0]:(i + 1) * B.shape[0], (i + 1) * B.shape[1]:(i + 2) * B.shape[1]] = C

A[((ny+2) - 1) * B.shape[0]:, ((ny+2) - 1) * B.shape[1]:] = np.eye(B.shape[0])

# plt.imshow(A, cmap='viridis')

print(A)
```

```python
##  Solution and source terms
u = np.zeros((nx+2)*(ny+2)) #Numerical solution
u_ex = np.zeros((nx+2)*(ny+2)) #Exact solution
F = np.zeros((nx+2)*(ny+2)) #Source term
```

```python
# Source term
def Source_int(x):
    return ((np.pi/xmax)**2+(np.pi/ymax)**2)*(np.sin(np.pi*x[0]/xmax)*np.sin(np.pi*x[1]/ymax))
def Source_bnd(x):
    return np.sin(np.pi*x[0]/xmax)*np.sin(np.pi*x[1]/ymax)
def Sol_sin(x):
    return np.sin(np.pi*x[0]/xmax)*np.sin(np.pi*x[1]/ymax)
for i in range(nptx):
    for j in range(npty):
        coord = np.array([i*hx,j*hy])
        u_ex[j*(nx+2) + i] = Sol_sin(coord)
    if i==0 or i==nptx-1: # Boundary x=0 ou x=xmax
        for j in range(npty):
            coord = np.array([i*hx,j*hy])
            F[j*(nx+2) + i]=Source_bnd(coord)
    else:
        for j in range(npty):
            coord = np.array([i*hx,j*hy])
            if j==0 or j==npty-1: # Boundary y=0 ou y=ymax
                F[j*(nx+2) + i]=Source_bnd(coord)
            else:
                F[j*(nx+2) + i]=Source_int(coord)
```

```python
#        
#Post-traintement u_ex+Visualization of the exct solution
uu_ex = np.reshape(u_ex,(nx+2 ,ny+2),order = 'F');
X,Y = np.meshgrid(xx,yy)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,uu_ex.T,rstride = 1, cstride = 1);
```

```python
#        
u = npl.solve(A,F)
uu = np.reshape(u,(nx+2 ,ny+2),order = 'F');
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X,Y,uu.T,rstride = 1, cstride = 1);
```

```python
# Error
print('norm L2 = ',npl.norm(u-u_ex))
```

```python
### Parameters

mu = 0.01 # Diffusion parameter
vx = 1 # Vitesse along x
# =============
```

```python
cfl = 0.2  # cfl =mu*dt/hx^2+mu*dt/hy^2 ou v*dt/h
dt = (hx**2)*(hy**2)*cfl/(mu*(hx**2 + hy**2)) # dt = pas de temps
dt = cfl*hx/vx
Tfinal = 0.8   # Temps final souhaitÃ©
```

```python
###### Matrice de Diffusion Dir/Neumann
B = scipy.sparse.diags([-mu/hx**2, mu*(2/hy**2 + 2/hx**2), -mu/hx**2], [-1, 0, 1], shape=(nx+2,nx+2), format = 'lil').toarray()
B[0,0] = 1
B[-1,-1] = 1
B[0,1]=0
B[-1,-2]=0
# plt.imshow(B, cmap='viridis')

C = -mu/hy**2*np.eye(nx+2)
C[0,0] = 0
C[-1,-1] = 0
# plt.imshow(C, cmap='viridis')


K2D = np.zeros(((ny+2) * B.shape[0], (ny+2) * B.shape[1]))
K2D[:B.shape[0], :B.shape[1]] = np.eye(B.shape[0])

for i in range(1,(ny+2)):
    K2D[i * B.shape[0]:(i + 1) * B.shape[0], i * B.shape[1]:(i + 1) * B.shape[1]] = B

for i in range((ny+2) - 2):
    K2D[(i + 1) * B.shape[0]:(i + 2) * B.shape[0], i * B.shape[1]:(i + 1) * B.shape[1]] = C

for i in range(1,(ny+2) - 1):
    K2D[i * B.shape[0]:(i + 1) * B.shape[0], (i + 1) * B.shape[1]:(i + 2) * B.shape[1]] = C

K2D[((ny+2) - 1) * B.shape[0]:, ((ny+2) - 1) * B.shape[1]:] = np.eye(B.shape[0])

# plt.imshow(A, cmap='viridis')

print(K2D)


#### Matrice de Convection  (Centré)
B = scipy.sparse.diags([-mu/hx**2 - vx/(2*hx), mu*(2/hy**2 + 2/hx**2),vx/(2*hx) -mu/hx**2], [-1, 0, 1], shape=(nx+2,nx+2), format = 'lil').toarray()
B[0,0] = 1
B[0,1]=0
B[-1,-1] = 3/(2*hx)
B[-1,-2] = -4/(2*hx)
B[-1,-3] = 1/(2*hx)
# plt.imshow(B, cmap='viridis')

C = -mu/hy**2*np.eye(nx+2)
C[0,0] = 0
C[-1,-1] = 0
# plt.imshow(C, cmap='viridis')


V2Dx = np.zeros(((ny+2) * B.shape[0], (ny+2) * B.shape[1]))
V2Dx[:B.shape[0], :B.shape[1]] = np.eye(B.shape[0])

for i in range(1,(ny+2)):
    V2Dx[i * B.shape[0]:(i + 1) * B.shape[0], i * B.shape[1]:(i + 1) * B.shape[1]] = B

for i in range((ny+2) - 2):
    V2Dx[(i + 1) * B.shape[0]:(i + 2) * B.shape[0], i * B.shape[1]:(i + 1) * B.shape[1]] = C

for i in range(1,(ny+2) - 1):
    V2Dx[i * B.shape[0]:(i + 1) * B.shape[0], (i + 1) * B.shape[1]:(i + 2) * B.shape[1]] = C

V2Dx[((ny+2) - 1) * B.shape[0]:, ((ny+2) - 1) * B.shape[1]:] = np.eye(B.shape[0])

# plt.imshow(A, cmap='viridis')

print(V2Dx)
```

```python
#### Global matrix : diffusion + convection
A2D = -(K2D + V2Dx) #-mu*Delta + V.grad
#
#
```
