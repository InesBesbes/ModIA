```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f_sin(t, T):
    return np.sin(2*np.pi*T*t)
```

```python
Fe = 8800
T = 440
```

```python
t = np.linspace(0,3,3*Fe)
```

```python
plt.plot(t,f_sin(t, T))
```

```python
from numpy import fft
```

```python
f = f_sin(t, T)
```

```python
N = len(f_sin(t, T))
```

```python
k = np.linspace(-N/2, N/2, 8800*3)
```

```python
plt.plot(k,abs(np.fft.fft(f_sin(t, T))))
```

```python
d = 1/Fe
f1 = np.fft.fftfreq(N, d)
```

```python
plt.plot(k,np.real(f1))
```

```python
plt.plot(k,np.imag(f1))
```

quand on varie d, la frquence change (l'chelle)

```python
f2 = np.fft.fftshift(f1)
```

```python
plt.plot(k,np.real(f2))
```

```python
plt.plot(k,np.imag(f2))
```

```python
plt.plot(f2,abs(f))
```

```python
def f_subplot(S):
    fig, axs = plt.subplots(2,2)
    s1 = np.fft.fft(S)
    axs[0, 0].plot(t, S)
    axs[0, 0].set_title('Signal')
    axs[0, 1].plot(k, np.abs(s1))
    axs[0, 1].set_title('Module fft')
    axs[1, 0].plot(k, np.real(s1))
    axs[1, 0].set_title('Re(fft)')
    axs[1, 1].plot(k, np.imag(s1))
    axs[1, 1].set_title('Im(fft)')

    plt.show()
```

```python
f_subplot(f)
```

```python
f_subplot(g)
```

```python
f_subplot(z)
```

```python
def gauss(x):
    return np.exp(-x**2)

def ind(x, I):
    r = np.zeros(len(x))
    for i in range(len(x)):
        if x[i]<= I[1] and x[i]>=I[0]:
            r[i]=1
        else: 
            r[i]=0
    return r
```

```python
def heaviside(x):
    r = np.zeros(len(x))
    for i in range(len(x)):
        if x[i]<0:
            r[i] = 0
        elif x[i]==0:
            r[i] = 1/2
        else:
            r[i] = 1
    return r
```

```python
t1 = np.linspace(-T,T)
plt.plot(t1,heaviside(t1))
```

```python
h1= np.fft.fft(heaviside(t1))
plt.plot(t1, abs(h1)**2)
```
