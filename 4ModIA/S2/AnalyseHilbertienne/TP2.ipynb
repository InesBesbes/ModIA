```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy import signal
```

```python
F1 = 440
F2 = 600
Fe = 2400

t1 = np.linspace(0,2,2*Fe)
t2 = np.linspace(0,3,3*Fe)
```

```python
plt.plot(t1,np.hanning(len(t1)))
plt.plot(t2,np.hanning(len(t2)))
```

```python
x = np.hanning(len(t1))
f, t, Sxx = signal.spectrogram(x)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
